### 여친 선물 - 워드 클라우드

```python
with open('Data/gift.txt','r',encoding='utf-8') as f:
    gift_text = f.read()
```

```python
tokens = okt.nouns(gift_text)
tokens[:10]
```

```python
# 영문자, 숫자 제거
import re           # regular expression
new_tokens = []
for token in tokens:
    text = re.sub('[A-Za-z0-9]','',token)   # 영대소문자와숫자를, ''지우겟다, token에서
    if text:                                # 빈텍스트가 있으면
        new_tokens.append(text)             # new_tokens에 추가를한다
```

```python
import matplotlib as mpl 
import matplotlib.pyplot as plt 
%matplotlib inline

# 한글폰트 사용
mpl.rc('font', family = 'Malgun Gothic')
mpl.rc('axes', unicode_minus=False)
```

```python
import nltk
```

```python
- 아래 두 셀을 반복해서 사용하여 
```

```python
gf_gift = nltk.Text(new_tokens, name='여친선물')
plt.figure(figsize=(15,6))
gf_gift.plot(50)
plt.show()
```

```python
# 위 데이터 중에서 쓸데없는 단어 지우기 (ex) 것,수,여친,선물,저,제 등) / 노가다로 하나하나 써야함
stop_words = []
new_tokens = [word for word in new_tokens if word not in stop_words]
new_tokens[:10]         # 앞에 10개만 뽑아서 확인해보기
```

### 워드 클라우드 그리기

```python
import numpy as np 
from PIL import Image
from wordcloud import WordCloud
```

```python
data = gf_gift.vocab().most_common(300)

wc = WordCloud(
    font_path='c:/Windows/Fonts/malgun.ttf',
    relative_scaling=0.2,
    background_color='white'
).generate_from_frequencies(dict(data))
```

```python
plt.figure(figsize=(12,6))
plt.imshow(wc)
plt.axis("off")
plt.show()
```
