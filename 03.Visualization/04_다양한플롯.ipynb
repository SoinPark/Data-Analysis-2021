### 1. Bar Chart

```python
y = [2, 3, 1]
x = np.arange(len(y))
xlabel = ['가', '나', '다']
plt.title("Bar Chart")
plt.bar(x, y)
plt.xticks(x, xlabel)
plt.yticks(sorted(y))
plt.xlabel("가나다")
plt.ylabel("빈도 수")
plt.show()
```

```python
np.random.seed(0)

people = ['몽룡', '춘향', '방자', '향단']
y_pos = np.arange(len(people))
performance = 3 + 10 * np.random.rand(len(people))
error = np.random.rand(len(people))

plt.title("Barh Chart")
plt.barh(y_pos, performance, xerr=error, alpha=0.4)
plt.yticks(y_pos, people)
plt.xlabel('x 라벨')
plt.show()
```

```python
# 타이타닉 데이터 - 선실별 승객수
import seaborn as sns
titanic = sns.load_dataset('titanic')
titanic.head(3)
```

```python
df = titanic[['class','pclass']].groupby('class').count()
df
```

```python
y = df.pclass.values
y
```

```python
x = np.arange(len(y))
xlabel = df.index.tolist()
xlabel
```

```python
plt.title('타이타닉 선실별 승객')
plt.bar(x, y)
plt.xticks(x, xlabel)
plt.xlabel("선실")
plt.ylabel("승객수")
plt.show()
```

### 2.파이 차트

```python
labels = ['개구리', '돼지', '개', '통나무']
sizes = [15, 30, 45, 10]
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']
explode = (0, 0.1, 0, 0)
plt.title("Pie Chart")
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=90)
plt.axis('equal')
plt.show()
```

### 4. 산점도(Scatter plot)
- 두 개의 실수 데이터 집합의 상관관계를 살펴볼 때 사용

```python
np.random.seed(0)
X = np.random.randn(100)
Y = np.random.randn(100)
plt.title("Scatter Plot")
plt.scatter(X, Y)
plt.show()
```

```python
N = 30
np.random.seed(0)
x = np.random.rand(N)
y1 = np.random.rand(N)
y2 = np.random.rand(N)
y3 = np.pi * (15 * np.random.rand(N))**2
plt.title("Bubble Chart")
plt.scatter(x, y1, c=y2, s=y3)
plt.show()
```

- Anscombe's Quartet

```python
ans = sns.load_dataset('anscombe')
ans.shape
```

```python
ans.head()
```

```python
ans.tail()
```

```python
ans.groupby('dataset').describe()
```

```python
ans[ans.dataset == 'I']
```

```python
fig, axes = plt.subplots(2, 2)
for index, roman in enumerate(['I','II','III','IV']):
    ds = ans[ans.dataset == roman]
    ax = axes[index//2, index%2]
    ax.scatter(ds.x, ds.y)

plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(10,8), sharex=True, sharey=True)
for index, roman in enumerate(['I','II','III','IV']):
    ds = ans[ans.dataset == roman]
    ax = axes[index//2, index%2]
    ax.scatter(ds.x, ds.y)

plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(10,8), sharex=True, sharey=True)
for index, roman in enumerate(['I','II','III','IV']):
    ds = ans[ans.dataset == roman]
    ax = axes[index//2, index%2]
    ax.scatter(ds.x, ds.y)
    ax.plot([3, 20], [4.5, 13], 'r-', lw=2)     # y = 0.5 * x + 3
    ax.set_title(f'Dataset {roman}')

fig.suptitle("Anscombe's Quartlet", fontsize = 16)
plt.tight_layout()
plt.show()
```
