## 여러개의 그림 그리기

```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.rcParams["font.family"] = 'NanumBarunGothic'
plt.rcParams["axes.unicode_minus"] = False
```

- Figure 객체
- 그림의 크기를 결정할 떄 사용 / 자주는 안쓴다

```python
np.random.seed(2021)
plt.figure(figsize=(10,2))
plt.title("figure size : (10,2)")
plt.plot(np.random.rand(100))
plt.grid()
plt.show()
```

- Axes 객체와 subplot 명령
- subplot(행, 열, 원하는 위치값)

```python
subplot(2, 1, 1)    # 2행 1열의 첫번째, 즉 윗부분 플롯
subplot(2, 1, 2)    # 2행 1열의 두번째, 즉 아랫부분 플롯
```

```python
x1 = np.linspace(0.0, 5.0)      # 디폴트값 50개!!!! / 0~50 개
x2 = np.linspace(0.0, 2.0)
y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)   # cosine/ 'sin파이x값' 이라고 나타냄
y2 = np.cos(2 * np.pi * x2)
```

```python
# 그림 그리기
ax1 = plt.subplot(2, 1, 1)
plt.title("2 Subplot")
ax1.plot(x1, y1, 'yo-')     # yellow에 동그라미에 실선으로 표시한다
ax1.set_ylabel('Damped oscillation')    # ax1으로 바뀌면서 set_ylabel로됨

ax2= plt.subplot(2, 1, 2)
ax2.plot(x2, y2, 'r.-')     # red에 점에 실선으로 표시한다
ax2.set_ylabel('Undamped')

plt.xlabel('time (sec)')
plt.show()
```

```python
# 위와같은 다른방법 위와같은 다른방법
ax1 = plt.subplot(211)
ax1.plot(x1, y1, 'yo-')
plt.title("2 Subplot")    
ax1.set_ylabel('Damped oscillation') 

ax2= plt.subplot(212)
ax2.plot(x2, y2, 'r.-')   
ax2.set_ylabel('Undamped')
plt.xlabel('time (sec)')

plt.tight_layout()
ax1.grid(), ax2.grid()      # plt.grid보다 ax1.grid 처럼 지정해줘서 출력하는게 정확!
plt.show()
```

```python
np.random.seed(0)

plt.subplot(221)
plt.plot(np.random.rand(5))
plt.title("axes 1")

plt.subplot(222)
plt.plot(np.random.rand(5))
plt.title("axes 2")

plt.subplot(223)
plt.plot(np.random.rand(5))
plt.title("axes 3")

plt.subplot(224)
plt.plot(np.random.rand(5))
plt.title("axes 4")

plt.tight_layout()
plt.show()
```

```python
# 위보다 간단한 방법
fig, axes = plt.subplots(2, 2)

np.random.seed(0)
axes[0, 0].plot(np.random.rand(5))
axes[0, 0].set_title("axes 1")
axes[0, 1].plot(np.random.rand(5))
axes[0, 1].set_title("axes 2")
axes[1, 0].plot(np.random.rand(5))
axes[1, 0].set_title("axes 3")
axes[1, 1].plot(np.random.rand(5))
axes[1, 1].set_title("axes 4")

plt.tight_layout()
plt.show()
```

```python
# 위 표를 반복문 사용하여 나타내기

np.random.seed(0)

fig, axes = plt.subplots(2, 2)
for i in range(4):
    ax = axes[i//2, i%2]       # 행 인덱스 - i//2, 열 인덱스- i%2
    ax.plot(np.random.rand(5))
    ax.set_title(f"axes {i+1}")
    ax.grid()

plt.tight_layout()
plt.show()
```
